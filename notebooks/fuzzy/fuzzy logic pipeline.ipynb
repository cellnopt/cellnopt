```python
%pylab inline
from cno import CNORfuzzy, cnodata
```

```python
c = CNORfuzzy(cnodata("PKN-ToyPCB.sif"), 
                        cnodata("MD-ToyPCB.csv"))
```

```python
# you may want to pre-process the data
# See the doc of preprocessing for details
c.preprocessing(expansion=True) # expansion is True by default
```

```python
c.plot_model()
# c._model.svg would give better resolution
```

```python
c.optimise(N=6, maxtime=10)
```

```python
# Looking at the fitness of the 5 runs, it seems to converge nicely
c.plot_fitness()
```

```python
# this figure tells you the mse as a function of number of parameters
# The x-axis give the threshold 
# so, here, the threshold should be 10-2 before the blue line(MSEs) 
# increases
c.plot_mses()
```

```python
c.plot_errors(0.01) # 0.01 is chosen based on previous image
```

```python
c.report()
```
