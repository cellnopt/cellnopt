You may simply use the **add_edge** providing the link (+ or -)

```python
c.add_edge("B", "d", link='+')
# equivalent to 
c.add_reaction("B=d")
```

## Setting attributes on nodes and edes

```python
c = CNOGraph()
c.add_reactions(['A=B', 'B=C', 'C=E1', 'C=E2'])
c.graph_options['graph']['rankdir'] = 'LR'
c.png
```

```python
c = CNOGraph()
c.add_reactions(['A=B', 'B=C', 'C=E1', 'C=E2'])
colors = {'A':1, 'B':.9, 'C':.8, 'E1':0.4, 'E2':0.4}
for k,v in colors.iteritems():
    c.node[k]['flux'] = v
c.plot(node_attribute='flux', cmap='gist_heat', colorbar=True)
```

```python
# similarly for edges:
c = CNOGraph()
c.add_reactions(['A=B', 'B=C', 'C=E1', 'C=E2'])
flux = {'A':1, 'B':.9, 'C':.8, 'E1':0.4, 'E2':0.4}
for k,v in flux.iteritems():
    c.node[k]['flux'] = v
for edge in c.edges():
    # color edge with flux input node
    e1,e2 = edge[0],edge[1]
    c.edge[e1][e2]['flux'] = flux[e1]
    c.edge[e1][e2]['label'] = "delta " + str(flux[e1] - flux[e2])
    
c.plot(node_attribute='flux', edge_attribute='flux', cmap='gist_heat', 
       colorbar=True)
```

## More on creating your own graph

```python
c2 = CNOGraph()
```

```python
c2.add_reaction("A=C")
c2.add_reaction("A=B")
c2.add_reaction("C+!B=D")
c2.png
            
```

```python
c2.expand_and_gates()
c2.png
```

```python
if ('C','D') in c2.edges(): c2.remove_edge('C', 'D')
if ('B', 'D') in c2.edges():c2.remove_edge('B', 'D')
c2.png
```
