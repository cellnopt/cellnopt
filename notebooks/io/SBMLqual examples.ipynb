# SBMLqual

cellnopt provide some tools to read/write SBMLqual format. IT does not rely on libsbml but should
be able to read/write logic models. 

```python
%pylab inline
from cno.io import sbmlqual
```

### Reading a known logic model from BioModel

```python
from bioservices import BioModels
```

```python
b = BioModels()
```

```python
models = b.getModelsIdByName('macnamara')
models
```

```python
model = b.getModelSBMLById(models[0])
```

```python
with open('test.xml','w') as fh:
    fh.write(model.encode('utf-8'))
```

```python
# the reader creates a SIF file out of the SBML
reader = sbmlqual.SBMLQual()
sif = reader.read_sbmlqual("test.xml")
```

```python
# data is not stored in BioModel, but in cellnopt
from cno import CNOGraph, cnodata, XMIDAS
data = XMIDAS(cnodata("MD-ToyPB.csv"))
```

```python
c = CNOGraph(sif, data)
```

```python
c.plot(show=False, filename='test.png')
```

### Creating the SBMLqual

```python
from cno import CNOGraph, cnodata, XMIDAS
data = XMIDAS(cnodata("MD-ToyPB.csv"))
sif = cnodata("PKN-ToyPB.sif")

c1 = CNOGraph(sif, data)
```

```python
c1.to_sbmlqual("PKN-ToyPB.xml")
```

### Reading back

```python
c2 =  CNOGraph("PKN-ToyPB.xml", data)
```

```python
c1 == c2
```
