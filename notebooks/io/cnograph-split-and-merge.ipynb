# Example of split_node and merge_node functionalities (cellnopt.core.cnograph)

```python
from cno.io import CNOGraph
%pylab inline
```

```python
c = CNOGraph()
```

```python
c.add_reaction("A+B=C")
c.add_reaction("C=D")
c.add_reaction("C=E")
c.add_reaction("D^E=F")
```

```python
c.svg
```

```python
c.split_node("C", ["C1", "C2"])
c.split_node("E", ["E1", "E2"])
c.svg
```

```python
c.merge_nodes(["C1", "C2"], "C")
c.svg
```

```python
# and back to the original network:
c.merge_nodes(["E1", "E2"], "E")
```
