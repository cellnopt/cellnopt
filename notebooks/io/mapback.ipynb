# MapBack Tutorial
The idea of the mapback class is to map a subgrap from a processed model onto the original network. 

```python
%pylab inline
matplotlib.rcParams['figure.figsize'] = (8,6)
```

```python
from cno import CNOGraph, cnodata, cnograph
from cno.io.mapback import MapBack
reload(cnograph)
```

### The Prior knowledge network 

```python
pknmodel = cnograph.CNOGraph(cnodata("PKN-ToyMMB.sif"), cnodata("MD-ToyMMB.csv"))
pknmodel.plot(filename='pknmodel.png')
```

### The processed network 

```python
model = cnograph.CNOGraph(cnodata("PKN-ToyMMB.sif"), cnodata("MD-ToyMMB.csv"))
model.preprocessing()
model.png
```

### The optimised network (subgraph of the processed one)

Let us set the style of the optimiesd network to be 0 if an edge is not present and 1 if present and plot the optimised network

```python
# in the Network, to add a style, we should use the edges (for now)
optimised = model.copy()
reactions = [reaction for reaction in optimised.reactions 
             if 'Mek' in reaction]
for edge in optimised.edges():
    if 'Mek' in edge[0] or 'Mek' in edge[1]:
        optimised.edge[edge[0]][edge[1]]['optimised'] = 1
    else:
        optimised.edge[edge[0]][edge[1]]['optimised'] = 0
optimised.plot(edge_attribute='optimised')

```

### The mapback class

```python
# artefact from previous cell where we modified the model
mp = MapBack(pknmodel, model)
```

```python
print reactions
```

```python
mp.mapback(reactions)
```

A convenient way to visualise and check the results is to use the plotall 
method:

```python
mp.plotall(reactions)
```
