# Example of CNORfeeder link inference

In this example, we use **cno.feeder** to show how to infer missing links in the PKN (from the data). To do so we use **cno.feeder** that calls **CNORfeeder** under the hood.

The visualisation is done with cellnot.core, which should be integrated in **cno** package asap

Author: Thomas Cokelaer, Sept 2014

```python
from cellnopt.core import cnodata, CNOGraph, cnograph
from cno import Feeder

%pylab inline
matplotlib.rcParams['figure.figsize'] =  (10.0, 6.0)

matplotlib.rcParams['figure.dpi'] =  160

```

```python
feeder = Feeder()
```

```python
feeder.run(cnodata("PKN-ToyMMB.sif"), cnodata("MD-ToyMMB.csv"), k=4, verbose=False)
```

```python
feeder.newlinks
```

```python
reload(cnograph)
c = cnograph.CNOGraph(cnodata("PKN-ToyMMB.sif"), cnodata("MD-ToyMMB.csv"))

# TEMPORARY solution to emphasize the new links that are proposed by CNORfeeder
for e in c.edges():
    c.edge[e[0]][e[1]]['ecolor'] = 0
for link in feeder.newlinks: 
    a,b = link.split("=")
    c.add_edge(a,b,link='+', ecolor=.5, penwidth=3) # all positive here by chance
    c.edge[a][b]['penwidth'] = 5    
    
c.plot(edge_attribute='ecolor', edge_attribute_labels=False, cmap='jet')
#c.plot()
```
